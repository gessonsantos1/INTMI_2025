```python
salario = float(input("Digite o salário atual: R$ "))

if salario <= 1500:
    aumento = 0.10
elif salario <= 3000:
    aumento = 0.07
else:
    aumento = 0.05

novo_salario = salario * (1 + aumento)

print(f"Novo salário: R$ {novo_salario:.2f}")
```

```python
soma = 0
contador = 0

while True:
    idade = int(input("Idade (negativa para parar): "))
    if idade < 0:
        break
    soma += idade
    contador += 1

if contador > 0:
    print("Média das idades:", soma / contador)
else:
    print("Nenhuma idade válida digitada.")
```

```python
pos = neg = zero = 0

for _ in range(10):
    n = float(input("Número: "))

    if n > 0:
        pos += 1
    elif n < 0:
        neg += 1
    else:
        zero += 1

print("Positivos:", pos)
print("Negativos:", neg)
print("Zeros:", zero)
```
