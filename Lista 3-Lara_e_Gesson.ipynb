```python
#17. Dada uma matriz 3x3 com valores reais, calcule o determinante (sem usar bibliotecas externas).
matriz = []

print("Digite os elementos da matriz 3x3 (valores reais):")

for i in range(3):
    linha = []
    for j in range(3):
        valor = float(input(f"Elemento [{i+1},{j+1}]: "))
        linha.append(valor)
    matriz.append(linha)

a, b, c = matriz[0]
d, e, f = matriz[1]
g, h, i = matriz[2]

determinante = a*(e*i - f*h) - b*(d*i - f*g) + c*(d*h - e*g)

print("\nDeterminante da matriz:", determinante)
```

```python
#18. Faça um programa que leia duas matrizes 3x3 e calcule a matriz soma e a matriz diferença.
def ler_matriz(nome):
    matriz = []
    print(f"Digite os elementos da matriz {nome} (3x3):")
    for i in range(3):
        linha = []
        for j in range(3):
            valor = float(input(f"Elemento [{i+1},{j+1}]: "))
            linha.append(valor)
        matriz.append(linha)
    return matriz

def soma_matrizes(A, B):
    return [[A[i][j] + B[i][j] for j in range(3)] for i in range(3)]

def subtrai_matrizes(A, B):
    return [[A[i][j] - B[i][j] for j in range(3)] for i in range(3)]

matriz_a = ler_matriz("A")
matriz_b = ler_matriz("B")


matriz_soma = soma_matrizes(matriz_a, matriz_b)
matriz_diferenca = subtrai_matrizes(matriz_a, matriz_b)


print("\nMatriz A:")
for linha in matriz_a:
    print(linha)

print("\nMatriz B:")
for linha in matriz_b:
    print(linha)

print("\nMatriz Soma (A + B):")
for linha in matriz_soma:
    print(linha)

print("\nMatriz Diferença (A - B):")
for linha in matriz_diferenca:
    print(linha)
```

```python
#19. Solicite dois conjuntos de 10 elementos fornecidos pelo usuário. Mostre a interseção, união e diferença (A - B e B - A).
conjunto_a = set()
print("Digite 10 elementos para o conjunto A:")
while len(conjunto_a) < 10:
    elemento = input(f"Elemento {len(conjunto_a)+1}: ")
    conjunto_a.add(elemento)

conjunto_b = set()
print("Digite 10 elementos para o conjunto B:")
while len(conjunto_b) < 10:
    elemento = input(f"Elemento {len(conjunto_b)+1}: ")
    conjunto_b.add(elemento)

intersecao = conjunto_a & conjunto_b

uniao = conjunto_a | conjunto_b

diferenca_a_b = conjunto_a - conjunto_b
diferenca_b_a = conjunto_b - conjunto_a

print("\nConjunto A:", conjunto_a)
print("Conjunto B:", conjunto_b)
print("Interseção (A ∩ B):", intersecao)
print("União (A ∪ B):", uniao)
print("Diferença (A - B):", diferenca_a_b)
print("Diferença (B - A):", diferenca_b_a)
```
